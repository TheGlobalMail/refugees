```python
def top_countries(year):
    d = year.groupby('destination')
    o = year.groupby('origin')
    top_dest = d.persons_applied.sum().idxmax()
    top_origin = o.persons_applied.sum().idxmax()
    top_dest_num = d.get_group(top_dest).persons_applied.sum()
    top_origin_num = o.get_group(top_origin).persons_applied.sum()
    
    return [[top_dest, top_dest_num], [top_origin, top_origin_num]]
```

```python
# top dest, origin w/ numbers, by year
for n, r in years:
    print (n, top_countries(r))
```

```python
country_names_map = {
    "Iran (Islamic Republic of)": "Iran",
    "Russian Federation": "Russia",
    "Syrian Arab Republic": "Syria",
    "Viet Nam": "Vietnam",
    "Democratic Republic of the Congo": "DR Congo",
    "Serbia (and Kosovo: S/RES/1244 (1999))": "Serbia (Incl Kosovo)",
    "United Republic of Tanzania": "Tanzania",
    "Venezuela (Bolivarian Republic of)": "Venezuela",
    "The former Yugoslav Republic of Macedonia": "Macedonia",
    "Republic of Moldova": "Moldova",
    "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
    "Libyan Arab Jamahiriya": "Libya",
    "C\xcc\xabte d'Ivoire": "Cote d'Ivoire",
    "C\xc3\xb4te d'Ivoire": "Cote d'Ivoire",
    "C\xf4te d'Ivoire": "Cote d'Ivoire",
    "China, Macao SAR": "Macao",
    "China, Hong Kong SAR": "Hong Kong",
    "Hong Kong SAR, China": "Hong Kong",
    "Macao SAR, China": "Macao",
    "Democratic People's Republic of Korea": "North Korea",
    "Dem. People's Republic of Korea": "North Korea",
    "TFYR Macedonia": "Macedonia",
    "Cura\xe7ao": "Curacao",
    "United States of America": "United States"
}
```

```python
def get_name(name):
    if name in country_names_map.keys():
        return country_names_map[name]
    else:
        return name
    
df['destination'] = df['destination'].apply(lambda x: get_name(x))
df['origin'] = df['origin'].apply(lambda x: get_name(x))
pops['country'] = pops['country'].apply(lambda x: get_name(x))
```

```python
num_convert = [col for col in pops.columns if col not in ['country', 'continent', 'region', 'country_code']]
pops[num_convert] = pops[num_convert].replace('\s', '', regex=True).astype(int)
pops.continent = [s.title() for s in pops.continent]
```

```python
for o in top_origins:
    num_applicants = origins.get_group(o).persons_applied.sum()
    className = re.sub(r'(\s|\(|\))', '', o)
    print '<div class="origin" data-name="'+className+'"><span class="origin-name"><h4>' + o + '</h4></span><span class="origin-num">' + '{:,}'.format(int(num_applicants)) + '</span></div>'
```

```python
test_l = []
# top 10 origins for a given dest
for name, r in dests:
    orig = r.groupby('origin')
    dest_d = {}
    top_10 = orig.persons_applied.sum().order(ascending=False)[:10].index.tolist()
    dest_d = {
       'destination': name,
       'applicants': r.persons_applied.sum()
    }
    if name in pops.country.unique().tolist():
        dest_population = pops_d['2010'][name]
        dest_d['per1k'] = r.persons_applied.sum() / dest_population
        dest_d['population'] = dest_population
        dest_d['continent'] = pops_d['continent'][name]
        dest_d['region'] = pops_d['region'][name]
    else:
        dest_d['per1k'] = "NA"
        dest_d['continent'] = "NA"
        dest_d['region'] = "NA"
        dest_population = "NA"
    
    origins = []
    for o in top_10:
        get_origin = orig.get_group(o)
        orig_total = get_origin.persons_applied.sum()
        orig_d = {}
        orig_d['origin'] = o
        orig_d['applicants'] = orig_total
            
        orig_years = []

        applied_year = get_origin.groupby('year').persons_applied.sum()
        applied_year_d = applied_year.to_dict()
        
        for yr in range(2004, 2013):
            if yr not in applied_year_d.keys():
                applied_year_d[yr] = 0
                
        for k in applied_year_d.keys():
            year_d = {
                'year': k,
                'applicants': applied_year_d[k]
            }
            if dest_population != "NA":
                year_d['applicantsPer1k'] = applied_year_d[k] / pops_d[str(min(2010, int(k)))][name]
            else:
                year_d['applicantsPer1k'] = "NA"
            
            orig_years.append(year_d)
        
        orig_d['years'] = orig_years
        origins.append(orig_d)
        
    dest_d['origins'] = origins
    test_l.append(dest_d)
```

```python
pprint(test_l)
```

```python
with open('nested.json', 'wb+') as fo:
    outjson = json.dumps(test_l)
    fo.write(outjson)
    fo.close()
```
